```
from datetime import datetime, timedelta
import os

# Content for the ICS file
ics_content = [
    "BEGIN:VCALENDAR",
    "VERSION:2.0",
    "PRODID:-//ExecuVoice//Launch Plan//EN",
    "CALSCALE:GREGORIAN",
    "METHOD:PUBLISH",
]

# Helper to create an event
def create_event(summary, description, start_dt, attendees):
    dt_str = start_dt.strftime("%Y%m%d")
    # All events are all-day events for the checklist
    event = [
        "BEGIN:VEVENT",
        f"SUMMARY:ðŸš€ ExecuVoice: {summary}",
        f"DTSTART;VALUE=DATE:{dt_str}",
        f"DTEND;VALUE=DATE:{dt_str}",
        f"DESCRIPTION:{description}",
        "STATUS:CONFIRMED",
        "TRANSP:TRANSPARENT", # Shows as "Free" so it doesn't block your calendar, just appears at top
    ]
    for email in attendees:
        event.append(f"ATTENDEE;CN=ExecuVoice Admin;RSVP=FALSE:mailto:{email}")

    event.append("END:VEVENT")
    return event

# Configuration
emails = ["fanya.uxd@gmail.com", "e.fanya.young@gmail.com"]
year = 2025

# Milestones Dictionary (Month, Day, Title, Description)
milestones = [
    (11, 24, "Start Sprint: Competency Matrix", "Finalize the Matrix and System Prompt."),
    (11, 25, "Legal: Sign PIIAA", "Ensure IP assignment is signed for all contributors."),
    (11, 26, "Outreach: LinkedIn Recon", "Follow Will Hunsinger & Tiffany Siu. Do not DM yet."),
    (11, 28, "Tech: 'Hello World' Audio", "Test STT -> LLM -> TTS loop in console."),
    (12, 1, "Tech: Frontend Build", "Start React Audio Visualizer build."),
    (12, 2, "Sales: EMAIL TIFFANY SIU (First Round)", "Send Script B: 'Help your founders interview'."),
    (12, 2, "Sales: EMAIL 80TWENTY", "Send Script A: 'Boutique Agency Pilot'."),
    (12, 5, "Legal: Terms of Service", "Draft Data Ownership clauses."),
    (12, 8, "Tech: Grading Engine", "Implement JSON scoring logic."),
    (12, 9, "Legal: Trademark Filing", "File 'Intent to Use' for ExecuVoice."),
    (12, 11, "Sales: EMAIL RIVIERA PARTNERS", "Send Script A to Will Hunsinger."),
    (12, 15, "Tech: Interrupt Handling", "Ensure AI stops speaking when user interrupts."),
    (12, 17, "Testing: Pro User Test", "Have a Director-level friend test the system."),
    (12, 19, "Sales: Stripe Setup", "Create payment link for $500 Founder's Pack."),
    (12, 22, "Launch: Soft Launch (Inner Circle)", "Send to mentors and advisors."),
    (12, 29, "Launch: HARD LAUNCH", "Public LinkedIn Post. Final email to partners."),
    (12, 31, "DEADLINE: MVP Complete", "Goal: 5 Founder's Packs Sold."),
]

# Generate events
for month, day, title, desc in milestones:
    date_obj = datetime(year, month, day)
    ics_content.extend(create_event(title, desc, date_obj, emails))

ics_content.append("END:VCALENDAR")

# Write to file
filename = "ExecuVoice_Launch_Calendar.ics"
with open(filename, "w") as f:
    f.write("\n".join(ics_content))

print(f"Successfully created {filename}")
print("Instructions: Open Google Calendar > Settings > Import & Export > Upload this file.")
```
